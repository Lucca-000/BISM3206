```python
# dictionary for future reference of DataFrame

import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt

# 1) Import CSV (or any format) into a DataFrame (must be in the folder to import)
df = pd.read_csv("your_file.csv")

# 2) Quick checks
df.head()        # first 5 rows
df.tail()        # last 5 rows
df.shape         # (rows, columns)
df.columns       # column names
df.dtypes        # data types
df.info()        # summary incl. non-null counts
df.describe()    # numeric summary stats
df.describe(include="all")  # includes non-numeric too

# 3) Common cleaning / prep
df.isna().sum()      # missing values per column
df.dropna()          # remove rows with missing values
df.fillna(0)         # fill missing values
df["col"].unique()   # unique values in a column
df["col"].value_counts()  # frequency table

# 4) Filtering and selecting
df["col"]                    # single column
df[["col1", "col2"]]         # multiple columns
df[df["col"] > 100]          # row filter
df.loc[0:4, ["col1"]]        # label-based select
df.iloc[0:5, 0:2]            # position-based select

# 5) Sorting and grouping
df.sort_values("col", ascending=False)
df.groupby("category")["value"].mean()


```
